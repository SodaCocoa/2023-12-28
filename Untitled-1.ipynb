```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from glob import glob

folder = r"C:\\Users\\bluecom011\Desktop\\python_1\\img\\"
fileList = glob(folder + '\\*')

imgList = []
for x in fileList:
    img = Image.open(x).resize((100, 200))
    imgArr = np.array(img)
    imgList.append(imgArr)

imgList = np.array(imgList)

# imgList를 1차원 배열로 평탄화
imgList = imgList.reshape(-1, 100, 200, 3)

# 예시로 첫 번째 이미지를 출력
plt.imshow(imgList[0])
plt.show()
```

```python
for cnt,i in enumerate(imgList):
    plt.subplot(1,len(imgList),cnt+1)
    plt.imshow(i)
```

```python
#리스트로
A반='3,가,나,다'.split(',')
B반='4,a-1,a-2,a-3'.split(',')
C반='5,b-1,b-2,a-3'.split(',')

dataList=[]
dataList.append(A반)
dataList.append(B반)
dataList.append(C반)
numList=[]

for i in dataList:
   # print(type(i[0])) 타입확인
    val=i[1]
    val=str(val)
    numList.append([])
    numList[-1].append(val) # -1은 numList의 마지막에 넣겠다.
    #print(i[0])
numList
```
