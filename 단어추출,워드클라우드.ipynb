```python
#위에 연설문중 가장 많이 나온 단어
from collections import Counter

def extract_most_common_words(text, num_words=5):
    # 소문자로 변환하고 특수 문자 제거
    cleaned_text = ''.join(char.lower() if char.isalpha() or char.isspace() else ' ' for char in text)

    # 단어 빈도 계산
    word_counts = Counter(cleaned_text.split())

    # 가장 많이 나온 단어 추출
    most_common_words = word_counts.most_common(num_words)

    return most_common_words

# 예시 텍스트
speech_text = """
Thank you for joining me today as we address a matter of utmost urgency and global significance – the issue of climate change.
The state of our planet demands our attention, collective action, and unwavering commitment to sustainable practices.
"""

# 가장 많이 나온 단어 추출 (상위 5개)
most_common_words = extract_most_common_words(speech_text, num_words=5)

# 결과 출력
print("Most common words:")
for word, count in most_common_words:
    print(f"{word}: {count}")
```

```python
#3글자이상단어 re.findAll없이 만든코드
from collections import Counter

def extract_most_common_words(text, num_words=5, min_word_length=3):
    # 소문자로 변환하고 특수 문자 제거
    cleaned_text = ''.join(char.lower() if char.isalpha() or char.isspace() else ' ' for char in text)

    # 단어 빈도 계산
    word_counts = Counter(cleaned_text.split())

    # 3글자 이상인 단어만 추출
    filtered_words = {word: count for word, count in word_counts.items() if len(word) >= min_word_length}

    # 가장 많이 나온 단어 추출
    most_common_words = Counter(filtered_words).most_common(num_words)

    return most_common_words

# 예시 텍스트
speech_text = """
Thank you for joining me today as we address a matter of utmost urgency and global significance – the issue of climate change.
The state of our planet demands our attention, collective action, and unwavering commitment to sustainable practices.
"""

# 가장 많이 나온 단어 추출 (상위 5개, 3글자 이상)
most_common_words = extract_most_common_words(speech_text, num_words=5, min_word_length=3)

# 결과 출력
print("Most common words (3 letters or more):")
for word, count in most_common_words:
    print(f"{word}: {count}")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd

# 주어진 텍스트
speech_text = """
today Thank you for joining our today as we address a matter of utmost urgency and global significance – the today issue of climate change.
The state of our planet demands our attention, collective action, and unwavering commitment to sustainable practices.
"""

# 워드클라우드 생성
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(speech_text)

# 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()

# 단어 빈도 막대그래프 생성
word_counts = pd.Series(wordcloud.words_)

# 막대그래프 시각화
plt.figure(figsize=(5, 3))
word_counts.sort_values().plot(kind='barh', color='green')
plt.title('Word Frequency in Speech')
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.show()
```
